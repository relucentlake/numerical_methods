```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = 0.01
x = np.arange(-1, 2+t, t)
i = (np.sin(2)*np.exp(2)-np.cos(2)*np.exp(2))/2-(np.sin(-1)*np.exp(-1)-np.cos(-1)*np.exp(-1))/2

def f(x):
    return np.exp(x)*np.sin(x)

def F(x):
    return np.sin(x)

def rectangular(x, f):
    F = 0
    for i in range(len(x)-1):
        F += f(x[i])*(x[i+1]-x[i])
    return F

def trapeze(x, f):
    F = 0
    for i in range(len(x)-1):
        F += (f(x[i+1])+f(x[i]))*(x[i+1]-x[i])/2
    return F

def Simpson(x, f):
    F = 0
    for i in range(len(x)-1):
        F += (f(x[i+1])+f(x[i])+4*f((x[i+1]+x[i])/2))*(x[i+1]-x[i])/6
    return F

print(i, 'интеграл')
print(rectangular(x, f), '- метод прямоугольников')
print(trapeze(x, f), '- метод трапеций')
print(Simpson(x, f), '- метод Симпсона')
```

```python
k = 1
step = []
error_r = []
error_t = []
error_S = []

for h in range(10):
    k /= 2
    j = np.arange(-1, 2+k, k)
    step.append(k)
    error_r.append(abs(i-rectangular(j, f)))
    error_t.append(abs(i-trapeze(j, f)))
    error_S.append(abs(i-Simpson(j, f)))

plt.loglog(step, error_r, '--', label = 'метод прямоугольников')
plt.loglog(step, error_t, ':', label = 'метод трапеций')
plt.loglog(step, error_S, label = 'метод Симпсона')
plt.legend()
plt.show()
```
