```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
```

```python
t = 0.1
x = np.arange(-3.1, 3.1, t)
r1 = range(1, len(x)-1)
r2 = range(2, len(x)-2)

f = np.arctan(x)
fd = 1/(1+x**2)
fdd = -2*x/((1+x**2)**2)

yf = np.array([(f[i+1]-f[i])/t for i in r1])
yc = np.array([(f[i+1]-f[i-1])/(2*t) for i in r1])
yc2 = np.array([(f[i+1]-2*f[i]+f[i-1])/t**2 for i in r2])
yc4 = np.array([(-f[i+2]+16*f[i+1]-30*f[i]+16*f[i-1]-f[i-2])/(12*t**2) for i in r2])
yd = np.array([fd[i] for i in r1])
ydd = np.array([fdd[i] for i in r2])
```

```python
xi = x[1:-1:] 
plt.plot(xi, yf, '--', label = 'правая разность')
plt.plot(xi, yc, ':', label = 'центральная разность')
plt.plot(xi, yd, label = 'производная')
plt.legend()
plt.show()
```

```python
xj = x[2:-2:]
plt.plot(xj, yc2, '--', label = '2 порядок')
plt.plot(xj, yc4, ':', label = '4 порядок')
plt.plot(xj, ydd, label = 'производная')
plt.legend()
plt.show()
```

```python
print(mean_squared_error(yd, yc))
print(mean_squared_error(yd, yf))
print(mean_squared_error(ydd, yc2))
print(mean_squared_error(ydd, yc4))
```
